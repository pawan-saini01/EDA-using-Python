Use numpy.where() to encode gender column.

```python
users['is_male'] = np.where(users.gender == 'M', 1, 0)
users1 = users.groupby(['occupation'])['is_male'].sum()/users.groupby(['occupation'])['gender'].count() * 100
users2 = pd.DataFrame(users1)
users2.rename(columns = { 0 : 'male_percent'}, inplace = True)
users2 = users2.reset_index(level = ['occupation'])
users3 = users2.sort_values(by = 'male_percent', ascending = False, ignore_index = True)
users3
```
